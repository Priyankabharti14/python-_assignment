```python
'''Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

1.Scatter Plot (sns.scatterplot):
     Use: Scatter plots are useful for visualizing the relationship between two continuous variables. They help
          identify patterns, trends, and potential outliers in the data.

2.Histogram (sns.histplot):
     Use: Histograms display the distribution of a single variable. They help you understand the frequency and
          spread of values within a continuous variable.
3.Box Plot (sns.boxplot):
    Use: Box plots are effective for visualizing the distribution of a dataset and identifying outliers. They
         display the median, quartiles, and potential outliers in a compact manner.
        
4.Bar Plot (sns.barplot):
    Use: Bar plots are suitable for comparing the mean or aggregate values of different categories. They help
         highlight differences or trends in categorical data.      
        
5.Pair Plot (sns.pairplot):
    Use: Pair plots are used for exploring relationships between multiple variables in a dataset. They provide a 
        matrix of scatterplots for each pair of variables and histograms along the diagonal'''     
```

```python
'''Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.
Note: timepoint, signal, event, and region are columns in the fmri dataset.'''

import seaborn as sns
fmri = sns.load_dataset('fmri')
sns.lineplot(x = 'timepoint', y = 'signal', data = fmri, style = 'event', hue = 'region')
```

```python
'''Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.'''

import seaborn as sns
import matplotlib.pyplot as plt
titanic = sns.load_dataset('titanic')
#first box plot for titanic 
plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
sns.boxplot(x='pclass', y='age', data=titanic)
plt.title("box plot for age ")
#second box plot for titanic
plt.subplot(1,2,2)
sns.boxplot(x='pclass', y='fare', data=titanic)
plt.title("box plot for fre")
plt.show()
```

```python
'''Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.'''

import seaborn as sns
import matplotlib.pyplot as plt

# Load the diamonds dataset
diamonds = sns.load_dataset('diamonds')

# Create a histogram for 'price' with 'cut' as the hue
plt.figure(figsize=(8, 4))
sns.histplot(data=diamonds, x='price', hue='cut', multiple='stack', kde=True)
plt.title('Histogram of Price with Cut as Hue')
plt.xlabel('Price')
plt.ylabel('Count')
plt.show()
```

```python
'''Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset.'''

import seaborn as sns
iris = sns.load_dataset('iris')
sns.pairplot(data=iris,hue='species')
```

```python
'''Que 6: Use the "flights" dataset from seaborn to plot a heatmap.'''

import seaborn as sns
import matplotlib.pyplot as plt

# Load the flights dataset
flights = sns.load_dataset('flights')

# Pivot the data to create a matrix for the heatmap
flights_pivot = flights.pivot_table(index='month', columns='year', values='passengers')

# Create a heatmap
plt.figure(figsize=(8, 4))
sns.heatmap(flights_pivot, annot=True, fmt='d', cmap='coolwarm', linewidths=.5)
plt.title('Passenger Count by Month and Year')
plt.show()
```
